```python
class Mul:
    def __init__(self,m):
        self.m = m
    
    def __call__(self,n): # Mul 클래스로 만든 객체에 인수를 전달해서 바로 호출할 수 있는 멤서드
        return self.m*n
    
    def mul(self,n):
        return self.m*n

if __name__ == "__main__":
    mul3 = Mul(3)
    mul5 = Mul(5)

    print(mul3(10))
    print(mul5(10))
```

```python
def mul(m): #여기서 받아온 m이 wrapper함수에 저장됨
    def wrapper(n):
        return m*n
    return wrapper # 함수가 함수를 리턴, 

if __name__ == "__main__":
    mul3 = Mul(3)
    mul5 = Mul(5)

    print(mul3(10))
    print(mul5(10))
```

```python
import time

def myfunc():
    start = time.time()
    print('함수가 실행됩니다')
    end = time.time()
    print(f'함수 수행시간: {end-start}초')

myfunc()

# 하지만 실행 시간을 측정해야 하는 함수가 myfunc 말고도 많다면 이런 코드를 모든 함수에 적용하는건 비효율적임
# 이때 클로저를 이용함
```

```python
import time

def elapsed(original_func):
    def wrapper():
        start = time.time
        result = original_func()
        end = time.time
        print(f'함수 수행시간: {end-start}초')
        return result
    return wrapper

@elapsed
def myfunc():
    print('함수가 실행됩니다')

# decorated_myfunc = elapsed(myfunc)
# decorated_myfunc()

myfunc()
```
